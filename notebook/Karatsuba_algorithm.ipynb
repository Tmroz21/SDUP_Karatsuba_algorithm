<h1 style="font-size:2rem;"> Karatshuba Algorithm </h1> 

<p>Algorytm Karatsuby to jeden z pierwszych przykładów zastosowania strategii „dziel i zwyciężaj” w celu przyspieszenia obliczeń arytmetycznych. Został opracowany w 1960 roku przez radzieckiego matematyka Anatolija Karacubę i opublikowany w 1962 roku. Było to znaczące odkrycie, które pokazało, że tradycyjne, „szkolne” metody mnożenia nie są optymalne i że możliwe jest wykonanie tych operacji znacznie szybciej, przynajmniej teoretycznie.</p>

<p>W czasie, gdy Karatsuba pracował nad swoim algorytmem, w Związku Radzieckim panowało duże zainteresowanie obliczeniami numerycznymi i efektywnością algorytmów, m.in. w związku z wyścigiem technologicznym i rozwojem komputerów. Młody Karatsuba miał wtedy około 23 lat i był studentem Andreja Kolmogorowa, jednego z najwybitniejszych matematyków XX wieku. To właśnie Kolmogorow zaproponował swoim studentom problem: czy można szybciej mnożyć liczby? Większość przypuszczała, że nie — Karatsuba pokazał, że można.</p>

<p>Algorytm opiera się na obserwacji, że da się zmniejszyć liczbę mnożeń potrzebnych do obliczenia iloczynu dwóch dużych liczb całkowitych. Tradycyjna metoda wymaga $n^2$ Karatsuba zaproponował rozbicie liczb na mniejsze fragmenty, wykonanie tylko trzech mnożeń zamiast czterech i rekonstrukcję końcowego wyniku za pomocą dodawań i przesunięć pozycyjnych.To podejście prowadzi do rekurencyjnej struktury algorytmu, w której każda warstwa dzieli problem na mniejsze podproblemy. Dzięki temu złożoność obliczeniowa spada do około $n^{log_2 3}$ co oznacza istotne przyspieszenie w porównaniu do klasycznej metody, zwłaszcza przy dużych liczbach.</p>

<h1 style="text-align:center">Porównanie złożoności</h1>

<p>Poniżej przedstawiony został wykres złożoności dla metody podstawowej oraz algorytmu Karatshuba</p>

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

def porownaj_zlozonosc(n):
    ns = np.arange(1, n + 1)
    klasyczna = ns**2
    karatsuba = ns**(np.log2(3))  # ≈ n^1.585

    plt.figure(figsize=(10, 6))
    plt.plot(ns, klasyczna, label="Klasyczne mnożenie (O(n²))", linewidth=2)
    plt.plot(ns, karatsuba, label="Karatsuba (O(n^log₂3))", linewidth=2, linestyle='--', color='orange')

    plt.title("Porównanie złożoności mnożenia")
    plt.xlabel("Długość liczby (n)")
    plt.ylabel("Złożoność (przybliżona liczba operacji)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
porownaj_zlozonosc(10)
```
